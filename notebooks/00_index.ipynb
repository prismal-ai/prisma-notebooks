```bash
# From the repo root
uv pip install -e ".[dev,all]"
echo "ANTHROPIC_API_KEY=sk-ant-..." > .env
```

## Regenerate the notebooks

These notebooks are generated automatically from `examples/*.py`. If you modify an example, run:

```bash
python notebooks/_generator/build_notebooks.py
python notebooks/_generator/build_index.py
```

The generator parses the module docstring, the section comments (`# ── Title ──` or 3-line blocks with dividers) and builds the corresponding `.ipynb` with alternating markdown/code.
